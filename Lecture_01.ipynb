###  File systems

The organization of computers is based on a _file system_. The file system is hierarchical, so at the top you'll find the _root directory_ or for Apple and Windows users, a _folder_. The root directory contains files and other folders which may also contain files and folders and etc. This continues, resulting in a tree of files and folders that make up the file system. The following figure is an example of a computer's file system:


````python
display.Image(filename='Figures/FileSystem.jpg')
````

You are probably familiar with the images like that to the left. The text to the right shows the exact same thing - but from your computer's viewpoint. 

Please note that _directory_ means the same thing as _folder_. It is just that in a graphical operating system with icons (left image), the term _folder_ makes sense - they look like folders. Whereas to the operating system, they are traditionally referred to as _directories_.

### Survival operating system commands

The operating system on a machine running Linux, Apple or Windows defines a large suite functions that can perform file system operations such as: list the contents of a folder (directory) or file; create a new directory; change the permission on files or directories; combine the contents of files, move files and directories, and so on. Whilst there will always exists "some operating system function to delete a file", at the very least it is likley to be called differently on each distinct operating system.  To make system-level operations independent of a particular operating system, Python has a built-in toolkit called "os" for operating system. We imported this in the first cell of this notebook (see `import os`) and now we will learn some of the things we use the `os` module for.  

#### Directory manipulations

Let's learn our first operating system command, which lists the contents of a directory, `os.listdir()`. This returns a list (not in any particular order) of all the things in the directory containing this notebook:

````python
os.listdir()
````

You can ignore anything with a '.' in front of it (`.git`, `.gitignore` and `.ipynb_checkpoints` in this example.)


Another useful command is `os.mkdir()` which creates a new directory.

````python
os.mkdir('MY_NEW_DIRECTORY')
````

To see if that worked, list the contents again:

````python
os.listdir()
````

And sure enough, there it is. The command `os.rmdir()` deletes a directory.

````python
os.rmdir('MY_NEW_DIRECTORY')
````

Make sure it was removed:  

````python
os.listdir()
````

Yup, it's gone.

To change the name of a directory to another name, use the command `os.rename()` (for change directory). 

````python
os.mkdir('TEMP')
print (os.listdir())
os.rename('TEMP','NEW_TEMP')
print (os.listdir())
os.rmdir('NEW_TEMP')
print (os.listdir())
````

#### File manipulations

Another handy thing is to view the contents of a file. To do this in Python, we use the command `open().readlines()`. This will spit the contents out for your viewing pleasure.

````python
open('Datasets/myfile.txt').readlines()
````

````python
contents = open('Datasets/myfile.txt').read()
output = open('newfile.txt', 'w') # open a file for writing - the 'w' denotes "write"
output.write(contents) # write the contents
output.close() # close the file
````

So what did we create? We created a copy of `myfile.txt` called `newfile.txt`. If you repeat the command, you will overwrite the existing output file.  

````python
open('newfile.txt').readlines()
````

To append to the end of a file, we use the `a` argument instead of `w` in the `open()` command  

````python
output = open('newfile.txt', 'a') # open a file for writing, and append if the file already exists
output.write(contents) # write the contents
output.close() # close the file
````

````python
open('newfile.txt').readlines()
````

To delete a file (analogous to deleting a directory), we use the command `os.remove()`.

````python
os.remove('newfile.txt')
````

### Command line Python scripts within Jupyter

As mentioned in the beginning of the lecture, you can run all the little programs you have been (and will be) writing, directly from the command line. Here's one way to do that that uses one of the many ["magic" commands] (https://ipython.readthedocs.io/en/stable/interactive/magics.html#cell-magics) that work with Jupyter notebooks. Our first is:  
```python
%%writefile PATH_TO_FILE.py
```
which writes the contents of a cell to the specified text file.  


Running this cell will place the contents of it (without the magic command) into a file in this directory called `hello.py`.  

````python
%%writefile ./hello.py
print("Hello World!")
````

Now you can run the program from your command line (after navigating to this directory) by typing:
```python
$ python hello.py
```
or from within this notebook:


````python
!python hello.py
````

Alternatively, you can use a different _magic_ command: `%run` to execute an external file:

````python
%run hello.py
````

````python
# Clean up after yourself. Remove the file we created.
os.remove('hello.py')
````
